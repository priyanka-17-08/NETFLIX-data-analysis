```python
df.drop_duplicates(inplace= True)
```

```python
df.dropna(subset = ['director', 'listed_in', 'date_added'], inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df.dtypes
```

```python
df.shape
```

```python
type_count= df['type'].value_counts()
print (type_count)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=type_count.index, y=type_count.values, hue=type_count.index, palette=['purple', 'skyblue'], legend=False)

plt.title('Distribution of Content by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['genres'] = df['listed_in'].apply(lambda x: x.split(','))
all_genres = sum(df['genres'], [])
genre_counts = pd.Series(all_genres).value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, hue=genre_counts.index, palette=['pink', 'purple', 'black', 'grey', 'lightblue','red','blue','orange','magenta','maroon'], dodge=False, legend=False)

plt.title('Most Common Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values,hue=top_countries.index, palette='Set3')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```
